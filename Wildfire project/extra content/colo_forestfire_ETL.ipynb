```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
file = "coloradoFiresGreaterThan50Acres.csv"
colofiresdb = pd.read_csv(file)
colofiresdb.head()
```

```python
for col in colofiresdb.columns: 
    print(col) 
```

```python
colofirestransdb = colofiresdb[['FIRE_NAME','SOURCE_REPORTING_UNIT_NAME','FIRE_YEAR','FIRE_SIZE','LATITUDE','LONGITUDE','STATE',]]


colorado_ff_db = pd.DataFrame(colofirestransdb)
colorado_ff_db
```

```python
connection_string = "postgres:postgres@localhost:5432/COForestFires"
engine = create_engine(f'postgresql://{connection_string}')

engine.table_names()
colorado_ff_db.to_sql(name='colofires', con=engine, if_exists='append', index=True)




```
